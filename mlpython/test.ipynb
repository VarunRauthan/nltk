```python
import pandas as pd
df = pd.read_excel("/home/varun/Downloads/SentimentAnalysisValidation_Nov212017(1).xlsx")
df.rename(columns={'Overall Support CSAT': 'CSAT'}, inplace=True)
df.drop(columns={'Surveyid for internal use (e.g. RI link)','Interaction Type','Response Date',})
df = df[['CSAT','Survey language','Full Text']]
df_english = df[(df['Survey language']=='English') | (df['Survey language']=='English (United Kingdom)')]
```

```python
import re
pd.options.mode.chained_assignment = None
df_english['Full Text'] = df_english['Full Text'].apply(lambda x : re.sub("[^a-zA-Z ' ' '.']"," ", re.sub("[0-9]","",x))).apply(lambda x: " ".join(x.split()))
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

#filtered_sentence(sent) = [w for w in format_sentence(sent) if not w in stop_words]

def format_sentence(sent):
    return({word: True for word in nltk.word_tokenize(sent) if not word in stop_words})

#filtered_sentence = [w for w in format_sentence(sent) if not w in stop_words]
```

```python
text = []
for index, row in df_english.iterrows():
    #text.append([format_sentence(row["Full Text"]),row["CSAT"]])
    text.append([format_sentence(row["Full Text"]),"low" if (row["CSAT"]>=0 and row["CSAT"]<6) else ("average" if (row["CSAT"]>5 and row["CSAT"]<9) else "good")])
text
```

```python
training = text#text[int((.9)*len(text)):]
```

```python
test = text[int((.1)*len(text)):]
```

```python
from nltk.classify import NaiveBayesClassifier

classifier = NaiveBayesClassifier.train(training)
```

```python
classifier.show_most_informative_features()
```

```python
example1 = "The representative was very helpful and informative."

print(classifier.classify(format_sentence(example1)))
```

```python
example2 = "Making excuses....Be honest on the date of delivery of money... Don't blame it to the bank. Fixed your system"

print(classifier.classify(format_sentence(example2)))
```

```python
from nltk.classify.util import accuracy
print(accuracy(classifier, test))
```
